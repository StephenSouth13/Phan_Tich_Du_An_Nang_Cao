```python
import yfinance as yf
import datetime as dt
import pandas as pd

# Thiết lập thời gian
start = dt.datetime(2016, 1, 1)
end = dt.datetime(2017, 12, 31)

# Lấy dữ liệu bằng yfinance (thay cho DataReader)
df = yf.download("AAPL", start=start, end=end)

# Lưu thành file CSV
df.to_csv('AAPL.csv')

# Đọc lại từ file CSV (Dùng pd.read_csv thay vì df.from_csv)
df_new = pd.read_csv('AAPL.csv', index_col=0, parse_dates=True)

# Hiển thị 5 dòng cuối
print(df_new.tail())
```

```python
import pandas as pd
import yfinance as yf # Thay thế cho pandas_datareader
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

# Cấu hình giao diện đồ thị
%matplotlib inline
sns.set_style('whitegrid') 

# 1. Thiết lập thời gian
end = dt.datetime.now()
start = dt.datetime(2016, 1, 1)

# 2. Lấy dữ liệu Stock bằng yfinance (Vô cùng ổn định)
df = yf.download("AAPL", start=start, end=end)

# 3. Lưu và Đọc dữ liệu đúng cách
df.to_csv('AAPL.csv')
# Dùng pd.read_csv để thay thế cho df.from_csv cũ
df = pd.read_csv('AAPL.csv', index_col=0, parse_dates=True)

# 4. Hiển thị dữ liệu cuối
print(df.tail())

# 5. Vẽ đồ thị
# Lưu ý: Cột dữ liệu của yfinance có thể là 'Adj Close' hoặc 'Close' tùy phiên bản
# Chúng ta dùng 'Close' (giá đóng cửa) cho phổ biến
plt.figure(figsize=(12, 8))
df['Close'].plot(legend=True)
plt.title('Apple Stock Price (AAPL)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```
